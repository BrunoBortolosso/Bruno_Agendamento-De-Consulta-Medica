## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
class Consulta:
    def __init__(self, paciente, medico, data):
        self.paciente = paciente
        self.medico = medico
        self.data = data
        self.__status = "Agendada"  # atributo privado

    # Mostra o status (getter)
    def get_status(self):
        return self.__status

    # Muda o status (setter)
    def set_status(self, novo_status):
        if novo_status in ["Agendada", "Confirmada", "Cancelada"]:
            self.__status = novo_status
        else:
            print("Status inv√°lido.")

    # Verifica se a consulta est√° pr√≥xima (urgente)
    def verificar_urgencia(self):
        from datetime import datetime
        hoje = datetime.now().date()
        data_consulta = datetime.strptime(self.data, "%Y-%m-%d").date()
        dias = (data_consulta - hoje).days
        if dias <= 1:
            return "Urgente"
        else:
            return "Normal"

    # Mostra os dados da consulta
    def exibir_detalhes(self):
        print(f"Paciente: {self.paciente}")
        print(f"M√©dico: {self.medico}")
        print(f"Data: {self.data}")
        print(f"Status: {self.__status}")
        print(f"Urg√™ncia: {self.verificar_urgencia()}")
class ConsultaOnline(Consulta):
    def __init__(self, paciente, medico, data, link):
        super().__init__(paciente, medico, data)  # chama a classe m√£e
        self.link = link

    # Mostra os dados + link da consulta online
    def exibir_detalhes(self):
        super().exibir_detalhes()
        print(f"Link da consulta: {self.link}")
```

## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

**Tarefas:**
- Consolidar todas as partes do projeto em um √∫nico script organizado.

- Testar todas as funcionalidades.

**C√≥digo:**

```python
# Importando bibliotecas
import sqlite3
from datetime import datetime



class Consulta:
    def __init__(self, paciente, medico, data):
        # Atributos principais
        self.paciente = paciente
        self.medico = medico
        self.data = data
        self.__status = "Agendada"  # Atributo privado (encapsulamento)

    # Getter: mostra o status
    def get_status(self):
        return self.__status

    # Setter: muda o status
    def set_status(self, novo_status):
        if novo_status in ["Agendada", "Confirmada", "Cancelada"]:
            self.__status = novo_status
        else:
            print("Status inv√°lido.")

    # M√©todo que verifica se a consulta est√° pr√≥xima (abstra√ß√£o)
    def verificar_urgencia(self):
        hoje = datetime.now().date()
        data_consulta = datetime.strptime(self.data, "%Y-%m-%d").date()
        dias = (data_consulta - hoje).days
        if dias <= 1:
            return "Urgente"
        else:
            return "Normal"

    # Mostra os dados da consulta
    def exibir_detalhes(self):
        print(f"Paciente: {self.paciente}")
        print(f"M√©dico: {self.medico}")
        print(f"Data: {self.data}")
        print(f"Status: {self.__status}")
        print(f"Urg√™ncia: {self.verificar_urgencia()}")



class ConsultaOnline(Consulta):
    def __init__(self, paciente, medico, data, link):
        super().__init__(paciente, medico, data)  # Usa o construtor da classe m√£e
        self.link = link

    # Polimorfismo: sobrescrevendo o m√©todo exibir_detalhes
    def exibir_detalhes(self):
        super().exibir_detalhes()
        print(f"Link da consulta: {self.link}")



# Cria a tabela no banco se n√£o existir
def criar_tabela():
    conexao = sqlite3.connect("consultas.db")
    cursor = conexao.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS consultas (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            paciente TEXT,
            medico TEXT,
            data TEXT,
            status TEXT
        )
    """)
    conexao.commit()
    conexao.close()

# Insere uma nova consulta no banco
def salvar_consulta(consulta):
    conexao = sqlite3.connect("consultas.db")
    cursor = conexao.cursor()
    cursor.execute("INSERT INTO consultas (paciente, medico, data, status) VALUES (?, ?, ?, ?)",
                   (consulta.paciente, consulta.medico, consulta.data, consulta.get_status()))
    conexao.commit()
    conexao.close()

# Lista todas as consultas do banco
def mostrar_consultas():
    conexao = sqlite3.connect("consultas.db")
    cursor = conexao.cursor()
    cursor.execute("SELECT * FROM consultas")
    dados = cursor.fetchall()
    for linha in dados:
        print(f"ID: {linha[0]} | Paciente: {linha[1]} | M√©dico: {linha[2]} | Data: {linha[3]} | Status: {linha[4]}")
    conexao.close()

# Deleta uma consulta pelo ID
def deletar_consulta(id_consulta):
    conexao = sqlite3.connect("consultas.db")
    cursor = conexao.cursor()
    cursor.execute("DELETE FROM consultas WHERE id = ?", (id_consulta,))
    conexao.commit()
    conexao.close()



# 1. Criar a tabela no banco
criar_tabela()

# 2. Criar uma consulta normal
consulta1 = Consulta("Ana Silva", "Dr. Jo√£o", "2025-05-24")
consulta1.set_status("Confirmada")
consulta1.exibir_detalhes()
salvar_consulta(consulta1)

print("\n--------------------\n")

# 3. Criar uma consulta online
consulta2 = ConsultaOnline("Lucas Lima", "Dra. Maria", "2025-05-25", "https://zoom.com/consulta123")
consulta2.set_status("Agendada")
consulta2.exibir_detalhes()
salvar_consulta(consulta2)

print("\n--------------------\n")

# 4. Mostrar todas as consultas salvas
print("Consultas no banco de dados:")
mostrar_consultas()

print("\n--------------------\n")

# 5. Deletar a consulta com ID 1 (teste)
print("Removendo consulta com ID 1...")
deletar_consulta(1)
print("Consultas atualizadas:")
mostrar_consultas()
```
